```python
import matplotlib.pyplot as plt
# see plots in jupyter notebook.
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)  # 'r-'
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
plt.show()
```

```python
# Multiplot in a canvas
plt.subplot(1,2,1)
plt.plot(x, y, 'r-')

plt.subplot(1,2,2)
plt.plot(y, x, 'b-')
```

```python
# object oriented method.
# create a figure object

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# axes.plot(x,y)
# plt.tight_layout() # fix layout

axes[0].plot(x,y)
axes[0].set_title('1st Plot')
axes[0].set_xlabel('x Label')
axes[0].set_ylabel('y Label')

axes[1].plot(y,x)
axes[1].set_title('2nd Plot')
axes[1].set_xlabel('x Label')
axes[1].set_ylabel('y Label')

plt.tight_layout()
```

```python
# Figure Size and DPI

fig = plt.figure(figsize=(3,2))  # 3 x 2 inches 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# plot in the same fig
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc='best') # loc = 0  is the best, by default
```

```python
fig=plt.figure()

ax= fig.add_axes([0,0,1,1])

#ax.plot(x,y, color='orange')
# ls  linestyle
# lw  linewidth
ax.plot(x,y, color='#9F84A3', lw=5, ls='--', 
        marker='o', markersize=20 , markerfacecolor='yellow',
                alpha=0.7) # RGB HEX code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])      
```

### Pandas Data Visualization

### Plot Types
There are several plot types built-in to pandas, most of them statistical plots by nature:

df.plot.area  
df.plot.barh  
df.plot.density  
df.plot.hist  
df.plot.line  
df.plot.scatter  
df.plot.bar  
df.plot.box  
df.plot.hexbin  
df.plot.kde  
df.plot.pie  
You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)

### Visualizing Time Series

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mcdon = pd.read_csv("./data/mcdonalds.csv", 
                   index_col='Date', parse_dates=True)
mcdon.head()
```

```python
mcdon['Adj. Volume'].plot(figsize=(12,4))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01', '2009-01-01'], ylim=(20,50))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01', '2009-01-01'], ylim=(20,50),
                        ls='--', c='red')
```

```python
import matplotlib.dates as dates
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01', '2009-01-01'], ylim=(20,50))
```

```python
# Edit time series ticks
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
stock=mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx, stock, '-')

# add grid lines
ax.yaxis.grid(True)
# ax.xaxis.grid(True)

# LOCATING
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATTING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

plt.tight_layout() # didnot work for time series x axes
fig.autofmt_xdate()
```
